```python
#Convierto el tipo de las columnas que estan en númerico y deberían estar como objeto
df['DOB']= pd.to_datetime(df['DOB'])
# Normalizar los valores de la columna 'HispanicLatino'
df['HispanicLatino'] = df['HispanicLatino'].str.capitalize()

df['EmpID'] = df['EmpID'].astype('str')
df['MarriedID'] = df['MarriedID'].astype('str')
df['MaritalStatusID'] = df['MaritalStatusID'].astype('str')
df['GenderID'] = df['GenderID'].astype('str')
df['EmpStatusID'] = df['EmpStatusID'].astype('str')
df['DeptID'] = df['DeptID'].astype('str')

df['LastPerformanceReview_Date'] = pd.to_datetime(df['LastPerformanceReview_Date'])

df["Position"] = df['Position'].replace('Data Analyst ','Data Analyst')
```

```python
#Miro si tengo valores duplicados en los ID de los empleados
duplicates_EmpID = df['EmpID'].duplicated()
duplicates_EmpID = duplicates_EmpID.sum()
print(f'Number of duplicate entries: {duplicates_EmpID}')
```

```python
df.groupby('Position')['PositionID'].count()
```

```python
filtered_df = df[(df['PositionID'] != 18) & (df['Position'] == 'Production Manager')]
print(filtered_df)
```

```python
#Creo una columna para mes, día y año
df[['Month', 'Day', 'Year']] = df['DateofTermination'].str.split('/', expand=True)
```

```python
#Convierto a fecha las columnas
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])
df['Antiquity'] = df['DateofTermination'] - df['DateofHire']
df['AntiquityInYears'] = df['Antiquity'].dt.days / 365
```

```python
condition_voluntarily = (df['EmploymentStatus'] == 'Voluntarily Terminated') 
condition_other = (df['EmploymentStatus'] != 'Active') & (df['EmploymentStatus'] != 'Voluntarily Terminated')
df_terminated = df[condition_voluntarily]
df_other = df[condition_other]
antiguedad_terminated = df_terminated['AntiquityInYears']
antiguedad_other_status = df_other['AntiquityInYears']
print(antiguedad_terminated.describe())
print(antiguedad_other_status.describe())
```

```python
# Contar el número de bajas en cada categoría
count_voluntary = df_terminated.shape[0]
count_non_voluntary = df_other.shape[0]

# Crear un gráfico de barras
labels = ['Voluntarily Terminated', 'Non-Voluntary Terminated']
counts = [count_voluntary, count_non_voluntary]

plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color=['brown', 'green'], edgecolor='black')
plt.xlabel('Termination Type')
plt.ylabel('Number of Employees')
plt.title('Comparison of Voluntary vs Non-Voluntary Terminations')
plt.show()
```

Decidismo analizar las bajas voluntarias ua que presentan un valor más elevado y en consecuencia provocan más coste para la empresa.

```python
#Creamos el histograma
plt.hist(antiguedad_terminated, bins=17, edgecolor='black', color='brown')

#Añadimos títulos y etiquetas
plt.title('Distribución de Bajas Voluntarias en Función de la Antigüedad')
plt.xlabel('Antigüedad en Años')
plt.ylabel('Número de Empleados')

#Mostramos el histograma
plt.show()
```

Tenemos niveles importantes de rotación tanto en personal nuevo con menos de 2 años de antigüedad en la empresa como en personal más antiguo como por ejemplo con más de 4 años.

Propuestas para el departamento de recursos humanos:
1- Mejorar la selección del personal, invertir más tiempo en las entrevistas, asegurar que la oferta de trabajo describe de forma fiable el puesto de trabajo.
2- Revisar el procedimiento de onboarding y de la formación inicial de los trabajadores. Asignar un tutor a los nuevos podría aumentar la tasa de permanencia y hacer el onboarding más ameno. 
3- Para el personal más antinguo elaboral planes de carrera y explorar formas de mantenerlos motivados. Analizar cada año los perfiles, detectar los que tienen potencial de crecimiento y realizar acciones de desarrollo a medida.

```python
#Creamos un gráfico con las bajas por año
order = sorted(df_terminated['Year'].unique())
plt.figure(figsize=(10, 3))
sns.countplot(data=df_terminated, x='Year', order=order, color='brown', edgecolor='black')
```

Las bajas de personal tienen una tendencia a la alza des de el 2010 hasta el 2016, alcanzando niveles máximos en el 2015 y 2016. En 2017 bajan y si miramos el contexto economico global, veníamos de una recesión. En 2018 vuelven a repuntar. 

```python
#Creamos un gráfico con las bajas por departamento
plt.figure(figsize=(10, 3))
sns.countplot(data=df_terminated, x='Department', color='brown',edgecolor='black')
```

```python
df_active = df[df['EmploymentStatus'] == 'Active']

# Contamos las bajas voluntarias por departamento
voluntary_terminations_by_department = df_terminated['Department'].value_counts()

# Contamos los empleados activos por departamento
active_employees_by_department = df_active['Department'].value_counts()

# Calculamos la proporción de bajas voluntarias respecto a los empleados activos + bajas de departamento
termination_ratio_by_department = voluntary_terminations_by_department / (active_employees_by_department+voluntary_terminations_by_department)

# Creamos un DataFrame para el gráfico
termination_ratio_df = termination_ratio_by_department.reset_index()
termination_ratio_df.columns = ['Department', 'Termination Ratio']
termination_ratio_df = termination_ratio_df.fillna(0)  # Reemplazar NaN con 0 para departamentos sin empleados activos

# Creamos el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=termination_ratio_df, x='Department', y='Termination Ratio', color='brown', edgecolor='black')
plt.xlabel('Department')
plt.ylabel('Termination Ratio')
plt.title('Voluntary Termination Ratio by Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

El departamento con el mayor número de bajas y ratio de rotación es el de producción.
Deberíamos prestar especial atención a este departamento para entender porque se va la gente puesto que la rotación tiene un coste elevado para la empresa.
Hints: realizar un benchmark salarial para identificar si los salarios de este personal en concreto esta acorde al mercado.

```python
#Creamos un gráfico con el promedio de satisfacción por departamento
order = df.groupby('Department')['EmpSatisfaction'].mean().sort_values().index

# Ajustamos el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Creamos el gráfico de barras agrupadas ordenado
sns.barplot(data=df, x='Department', y='EmpSatisfaction', estimator=np.mean, order=order,errorbar=None, color='brown', edgecolor='black')

# Añadimos etiquetas y título
plt.xlabel('Departamento')
plt.ylabel('Satisfacción Media')
plt.title('Satisfacción Media por Departamento')

# Mostramos el gráfico
plt.show
```

Aquí podemos ver que el departamento de producción no es el que tienen los mejores valores de satisfacción. De los 6 departamentos analizados, es el tercer menor scoring. 
Hint: coger la encuesta de satisfacción de donde han salido estos número y mirar que aspectos podemos mejorar para poder elevar esta puntuación.

```python
# Ordenamos los departamentos por el promedio de EngagementSurvey
order = df.groupby('Department')['EngagementSurvey'].mean().sort_values().index

# Ajustamos el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Creamos el gráfico de barras agrupadas ordenado
sns.barplot(data=df, x='Department', y='EngagementSurvey', estimator=np.mean, order=order, errorbar=None, color='brown', edgecolor='black')

# Añadimos etiquetas y título
plt.xlabel('Departamento')
plt.ylabel('Encuesta de Compromiso Media')
plt.title('Encuesta de Compromiso Media por Departamento')

# Mostramos el gráfico
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

El compromiso en el departamento con mayor rotación no es muy alto, la empresa debería bajar mejor la cultura a los empleados y mejorar los niveles de compromiso para poder esta forma reducir la rotación.
